```python
import requests
BASE = "http://localhost:5000"
print("Ready!")
```

```python
r = requests.get(f"{BASE}/health")
print(r.json())
```

```python
with open("datasets/coursea_data.csv", "rb") as f:
    r = requests.post(f"{BASE}/api/upload",
                      files={"file": ("coursea_data.csv", f, "text/csv")})

result = r.json()
print("Status:", r.status_code)
print("Rows:", result["row_count"])
print("Columns:")
for col in result["columns"]:
    print(f"  - {col['name']} ({col['dtype']})")
```
